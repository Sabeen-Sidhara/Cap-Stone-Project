# Support Vector Machine (SVM)

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('solar_panel_classification_data.csv')
```

```python
# Number of rows and Columns 
print("The Number of Row:",dataset.shape[0])
print("The Number of Column",dataset.shape[1])
```

```python
dataset.columns
```

```python
X1=dataset['Efficiency']
Y1=dataset['Degradation']
```

```python
# plot
```

```python
plt.plot(X1,marker='o',color='red',linestyle='dashed',
...      linewidth=1, markersize=3)
#plt.plot([1, 2, 3, 4], [1, 2, 3, 4],marker='o',color='BLUE',linestyle='dashed',
...      #linewidth=2, markersize=12)
plt.grid(True)#default
plt.grid(color='b', ls = '-.', lw = 1)
plt.title("Line plot",loc="right")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python

X, Y = np.meshgrid(X1, Y1)
#Z = np.sqrt(X**2 + Y**2)
Z=X-Y
cp=plt.contour(X, Y, Z)
plt.grid(True)#default
plt.grid(color='b', ls = '-.', lw = 1)
plt.title("Line plot",loc="right")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.colorbar(cp)
plt.show()
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')

X, Y = np.meshgrid(X1, Y1)
Z = np.sqrt(X**2 + Y**2)
cp=plt.contourf(X, Y, Z)
#ax.contour3D(X, Y, Z, 50, cmap='gray')
ax.plot_wireframe(X, Y, Z, color='black')
#ax.plot_surface(X,Y,Z,cmap='viridis', edgecolor='none')
plt.show()
```

```python
plt.boxplot([X1,Y1])
plt.show()
```

```python
plt.violinplot([X1,Y1])
```
