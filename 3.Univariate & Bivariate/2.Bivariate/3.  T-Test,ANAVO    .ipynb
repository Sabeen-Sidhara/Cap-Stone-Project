# Independent Samples -Unpaired T-Test 

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
other=dataset[dataset['Panel']=='Panel_A']['Efficiency']
central=dataset[dataset['Panel']=='Panel_A']['Degradation']
ttest_ind(other,central)
```

The P value is 0
T Score value tells that the group are different

The p-value of 0 supports the alternative  hypothesis (H1).

# ANOVA: Two Way Classification:

```python
y = dataset['output_category']
from sklearn.preprocessing import LabelEncoder, StandardScaler
label_encoder = LabelEncoder()
dep = label_encoder.fit_transform(y)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
anova_two=ols('dep ~ Efficiency + Degradation + Efficiency*Degradation',data=dataset).fit()
Two=sm.stats.anova_lm(anova_two,type=2)
Two
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
anova_two=ols('dep ~ Efficiency + Degradation +solar_irradiance+ Efficiency*Degradation*solar_irradiance',data=dataset).fit()
Two=sm.stats.anova_lm(anova_two,type=2)
Two
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
anova_two=ols('dep ~ temperature + humidity +wind_speed+cloud_cover+temperature*humidity*solar_irradiance*cloud_cover',data=dataset).fit()
Two=sm.stats.anova_lm(anova_two,type=2)
Two
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
anova_two=ols('dep ~ panel_age + panel_tilt_angle +latitude+longitude+panel_age*panel_tilt_angle*latitude*longitude',data=dataset).fit()
Two=sm.stats.anova_lm(anova_two,type=2)
Two
```
